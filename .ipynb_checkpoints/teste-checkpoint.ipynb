```python
from numpy import linspace, pi, sin, cos
from matplotlib.pyplot import plot, grid, show
```

```python
x = linspace(-2*pi,2*pi,100)
y = sin(x)
```

```python
grid('True')
plot(x,y)
show()
```

```python
grid(True)
plot(x,y)
show()
```
